```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing dataset
df=pd.read_csv("C:\\Users\\SAGAR\\Documents\\ashwini\\deep learning\\Complete-Deep-Learning-master\\Complete-Deep-Learning-master\\ANN\\Churn_Modelling.csv")
df
```

```python
X=df.iloc[:,3:13]
Y=df.iloc[:,13]
X
```

```python
#create dummy variable
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
gender
```

```python
geography
```

```python
#concatenate the data frames
X=pd.concat([X,geography,gender],axis=1)
```

```python
X
```

```python
X
```

```python
#Drop necessary column
X=X.drop(['Geography', 'Gender'], axis = 1)
```

```python
X
```

```python
#split dataset into training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler()
X_train = X_scaler.fit_transform(X_train)
X_test = X_scaler.transform(X_test)
X_test
```

```python
import keras 
from keras.models import Sequential #useful for creating neural network
from keras.layers import Dense#when you creating hidden layers
#from keras.layers import LeakyReLU,PReLU,ELU#activation function
from keras.layers import Dropout#is a regularization parameter,its used when your neural n/w are deep
```

```python
#intialising the ANN
classifier=Sequential()#these empty neural n/w currently,in that we know how many hidden layer be there and how many neuron be there
```

```python
#adding the input layer and the first hidden layer
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))#Dense layer to create my first hidden layer
classifier.add(Dropout(0.3))#1 para is o/p para is my 1st hidden layer and output dimension ,i am consider as 6 hidden neuron(unit =6),2 para is init,it is intilzation parameter.its indicated 
#how your weight need to be intialized.relu is activation function. i/p fetaure are 11,how many i/p feture are connected 
#to this hidden layer,menas your total column are 11 bcoz how many i/p fetaurer coonected to hidden layer.
#means when we creating neural n/w,your 1st layer i/p layer are 11 i/p neuron and 1st layer having 11 i/pneuron and 2nd is having 6 hidden neuron #use he uniform bcoz its work well.in 2nd layer having 6 hidden neuron and how i am coming to know that i have to add 6 .i have just put 6 now.later by using hyper optimization technique for to exact how many  neuron to used in hidden layer and how many hidden layer i have to used.  
```

```python
#adding second hidden layer and your 2nd layer having 6 hidden neuron,similarly 6 hidden neuron add in next layer
#classifier.add(Dense(output_dim=6,init='he_uniform',activation='relu'))
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))#init is weight initilztion parameter
classifier.add(Dropout(0.4))#when you put dropout then decrese accuracy but that not means it is having bad accuracy
#relu and lekeyrelu are most famous activation function.and relu prevent vanishing gredeient problem as comapred to sigmoid
```

```python
#adding output layer
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))#last layer is 1 dimentional(/1 hidden neuron) bcoz this is binary classification and based 
#on that be able to predict whether if it is grether than 0.5 the ans will be 1.that time i used sigmoid activation function.all your hidden layer should be having relu or leakyrelu.relu prevent vasishing gredient problem as compaired to sigmoid
```

```python
classifier.summary()#we can see that how many layer have,and how many output neuron are basically have and total number of parameter basically indicated. show how many weight and bias para have
```

```python
#compiling the ANN
classifier.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])#providinfg what kind of optimizer we r use,and what kind of loss function we can use
#and wht kind of is metrics r looking # in the 1st epoch weight are mulitly with i/p in each  and every layer after the each activation fu is getapplied
#and when u get final o/p getting it then you derive your loss fun and you create optimizer which will reduce your loss function.different typpes of optimizer @delta ,adagrad,rmsprop,adam, stoticstic gredient optimizer but currently adam is best optimizer
#loss = binary crossentropy bcoz when your o/p 0 and 1 you should prefer cross entropy if your o/p has multiple catogory then you should be use catogoricalcrossentopy and this info giving keras documenation.
```

```python
#fitting the ANN to the training set
model_history=classifier.fit(X_train,Y_train,validation_split=0.33,batch_size=10,nb_epoch=100)#validation_split bcoz so that  i can test my dataset/test model seprately for test dataset
#batch size used so that computation power become less and more amount of record you don't have to load at one time,your RAM are prteey much free,and able to run your program run on your laptop/ PC
#epoch used bcoz to incresing your accurcay it has come to 86 by using right kind of keral intializer which is your weight intializer parameter. training accuracy is become better for that reason used epoch .selecting right kind of intializer and by using simple neural network you get good acuracy,if you dont select parameter perfectly that time decrease your accuracy.validation accuracy shows means is require to select accurate parameter means kernel intializer is he uniform or other this para slection,if we select other para accurcay will be decreses
```

```python
#list all data in history
print(model_history.history.keys())
```

```python
#summarize history for accuracy
plt.plot(model_history.history.keys())
#summarize history for accuracy
plt.plot(model_history.history['acc'])
plt.plot(model_history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#summarize history for accuracy
plt.plot(model_history.history['loss'])
#summarize history for accuracy
plt.plot(model_history.history['val_loss'])
plt.title("model loss")
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#predicting the test set results#i try accuracy for test data and finally i will say which ever is greater than 0.5 consider data true /false
y_pred=classifier.predict(X_test)
y_pred=(y_pred>0.5)
y_pred
```

```python
#making the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
cm
```
