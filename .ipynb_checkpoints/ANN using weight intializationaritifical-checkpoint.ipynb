# Create Artificial neural n/w  using weight intialization tricks

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing dataset
df=pd.read_csv("C:\\Users\\SAGAR\\Documents\\ashwini\\deep learning\\Complete-Deep-Learning-master\\Complete-Deep-Learning-master\\ANN\\Churn_Modelling.csv")
df
```

```python
X=df.iloc[:,3:13]
Y=df.iloc[:,13]
X
```

```python
#create dummy variable
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
gender
```

```python
geography
```

```python
#concatenate the data frames
X=pd.concat([X,geography,gender],axis=1)
```

```python
X
```

```python
X
```

```python
#Drop necessary column
X=X.drop(['Geography', 'Gender'], axis = 1)
```

```python
X
```

```python
#split dataset into training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler()
X_train = X_scaler.fit_transform(X_train)
X_test = X_scaler.transform(X_test)
X_test
```

```python
import keras 
from keras.models import Sequential #useful for creating neural network
from keras.layers import Dense#when you creating hidden layers
#from keras.layers import LeakyReLU,PReLU,ELU#activation function
from keras.layers import Dropout#is a regularization parameter,its used when your neural n/w are deeep
```

```python
#intialising the ANN
classifier=Sequential()#these empty neural n/w currently,in that we know how many hidden layer be there and how many neuron be there
```

```python
#adding the input layer and the first hidden layer
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))#Dense layer to create my first hidden layer
classifier.add(Dropout(0.3))#1 para is o/p para is my 1st hidden layer and i am consider 6 hidden neuron(output =6),2 para is init,it is intilzation parameter.its indicated 
#how your weight need to be intialized.relu is activation function. i/p fetaure are 11,how many i/p feture are connected 
#to this hidden layer,menas your total column are 11 bcoz how many i/p fetaurer coonected to hidden layer.
#means when we creating neural n/w,your 1st layer i/p layer are 11 i/p neuron 
```

```python
#adding second hidden layer and your 2nd layer having 6 hidden neuron,similarly 6 hidden neuron add in next layer
#classifier.add(Dense(output_dim=6,init='he_uniform',activation='relu'))
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))#init is weight initilztion parameter
classifier.add(Dropout(0.4))#when you put dropout then decrese accuracy but that not means it is having bad accuracy
```

```python
#adding output layer
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))#last layer is 1 dimentional bcoz this is binary classification and based 
#on that be able to predict whether if it is grether than 0.5 the ans will be 1.that time i used sigmoid activation function.all your hidden layer should be having relu or leakyrelu.relu prevent vasishing gredient problem as compaired to sigmoid
```

```python
classifier.summary()#we can see that how many layer have,and how many out put neuron are basically have and show how many weight and bias para have
```

```python
#compiling the ANN
classifier.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])#providinfg what kind of optimizer we r use,and what kind of loss function we can use
#and wht kind of is metrics#weight are mulitly with i/p in each layer after the each activation fu is getapplied
#and when u get final o/p and dervie the loss fun and you create optimizer which reduce your loss function,like stoticstic gredient optimizer and adam is best optimizer
#loss when you o/p 0 and 1 you should prefer cross entropy if your o/p has ultiple catogory then you should be CPU catogoricalentopy and this documenation used in keras
```

```python
#fitting the ANN to the training set
model_history=classifier.fit(X_train,Y_train,validation_split=0.33,batch_size=10,nb_epoch=100)#validation split bcoz i can test my dataset sepratly for test dataset
#batch size used so that computation power become less and more amount of of data dont load at one time,your RAM are much free,
#epoch used to incresing your accurcay.traing accuracy ,validation accuracy shows means is require to select accurate parameter means kernel intializer is he uniform or other this para slection,if we select other para accurcay will be decreses
```

```python
#list all data in history
print(model_history.history.keys())
```

```python
#summarize history for accuracy
plt.plot(model_history.history.keys())
#summarize history for accuracy
plt.plot(model_history.history['acc'])
plt.plot(model_history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#summarize history for accuracy
plt.plot(model_history.history['loss'])
#summarize history for accuracy
plt.plot(model_history.history['val_loss'])
plt.title("model loss")
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#part 3 making prediction and evalute the model
```

```python
#predicting the test set results
y_pred=classifier.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
#making the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
```

```python
#calculate the accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,Y_test)
```
