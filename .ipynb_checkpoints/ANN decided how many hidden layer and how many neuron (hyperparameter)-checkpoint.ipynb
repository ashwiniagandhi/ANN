```python
# Artificial Neural Network#in this assignment is decided how many hidden layer and how many neuron in each and every hidden layer
#how to decide hidden layer an neuron

# Part 1 - Data Preprocessing

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('C:\\Users\\SAGAR\Documents\\ashwini\\deep learning\\Complete-Deep-Learning-master\\Complete-Deep-Learning-master\\ANN\\Churn_Modelling.csv')
dataset
#creaditscore to estimatedsalary are independent feature and exited coloumn are dependent feature
```

```python
X = dataset.iloc[:, 3:13]
y = dataset.iloc[:, 13]
X
#x and y are splited independent and feature are splited
```

```python
#Create dummy variables
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
geography#geography and gender feature are catogorical feature and convert this features
geography
```

```python
gender
```

```python
## Concatenate the Data Frames.concatinate geography,gender dataset with X and X is independent feature

X=pd.concat([X,geography,gender],axis=1)
X
```

```python
## Drop Unnecessary columns
X=X.drop(['Geography','Gender'],axis=1)
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
#X_train are 8000 record and X_test are 2000 record total record are 10000 
```

```python
# Feature Scaling#scale down all feature in same scale range.and easily back propogation happen feature scaling used for bcoz of find easily gredient descent ,conversion happen quickly,global minima point able to find easily 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Perform hyperparameter optimization
# Importing the Keras libraries and packages
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV#is uesd bcoz selecting right kind of hyper parameter that is called hyper parameter tunning
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,Activation,Embedding,Flatten,BatchNormalization
from keras.activations import relu,sigmoid 
```

```python
def create_model(layers,activation):#2 para provide and i can itrated this layer and create dense neuron/activation neuron
    model=Sequential()
    for i,nodes in enumerate(layers):#create dense neuron
        if i==0:         #i=0 means 1 st layer
            model.add(Dense(nodes,input_dim=X_train.shape[1]))#i/p dim provide i/p features
            model.add(Activation(activation))#which activation function applied into your hidden layer
        #aaply dropout rate
        else:
            model.add(Dense(nodes))#inside layer consist hidden neuron inhidden layer.inside layer consist list of hidden neuron
            model.add(Activation(activation)) #then applied activation function
            #applied dropout rate
    model.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))#in last layer(o/p layer) having 1 o/pAnd o/p as a 0 or 1,it isbinary o/p
    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])#in relu activation fun he uniform is bettern bcoz it work well with relu A.F
    return model
```

```python

model=KerasClassifier(build_fn= create_model,verbose=0)#build fun provide function name,
```

```python
layers=[[20],[40,20],[45,30,15],[30,20,10,3]]#consider this example,1st hidden layer having 20 neuron,in second hidden layer having in 2nd try to create 2 hidden layer 1st hidden layer having 40 neuron simily 2nd layer having 20 neuron.simliary 3 rd iteration
#u can used multiple iteration means multiple combination.#similary this combination it will create ,you can take many combination 1, combination define 1 neural n/w
activations=['sigmoid','relu']#2 a.f. are used for each combination .for all combination 1st iteration used sigmoid and 2nd used relu.1st combination having 20 neuron both activaton function applied on that .and determine which accuracy are better result.and that basically called best aprameterand best tuning hyper parameter .and finally 1 o/p will there of each combination 
param_grid=dict(layers=layers,activation=activations,batch_size=[126,254],epochs=[20])#created dictionry of parameter
grid2=GridSearchCV(estimator=model,param_grid=param_grid,cv=5)#provide grid search library#cross validation is 5,
```

```python
g_result=grid2.fit(X_train,Y_train)
```

```python
#check if we used 1 hidden layer/1 hidden neuron means 1st combination (above cell) then applied a.f. both a.f. are appplied and check which a.f. accurcy is better .if we used sigmoid for 1 hidden layer how it is accuracy,then check 2nd a.f. on same hidden layer combination.
#similary then check 2 nd combination which have 2 hiddenlayer and 2 hidden neuron ,applied both a.f. on this 2nd number combination and check both a.f. accuracy.which are bettern accuary , then used this a.f. and used this hidden layer which type of combination of hidden layer accuracy is better.
#same on 3 combination of hidden layers aaplied both a.f. and batch size.one time 1 a.f. and 1 batch size applied and check accuray for hidden ayer combinations.
#and from that we knows that how many hidden layer we used and what activation function useful and good.
```

```python
#model give best result#it will give which a.f. is best,which batch_size is best,how many layers are good how many hidden neuron is good
print(g_result.best_score_,g_result.best_params_)
```

```python
pred_y=grid2.predict(X_test)
y_pred=(pred_y>0.5)
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_pred,Y_test)
score=accuracy_score(y_pred,Y_test)
score
```
